```python
#Getting the US GDP data
import pandas as pd
data = pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/projects/coursera_project/clean_gdp.csv")
df = pd.DataFrame(data)
df.head()
```

```python
#Getting the US Unemployment Data
data2 = pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/projects/coursera_project/clean_unemployment.csv")
dfb = pd.DataFrame(data2)
dfb.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
year = df['date']
year
```

```python
unemployment = dfb['unemployment']
unemployment
```

```python
gdp_data = df['change-current']
gdp_data
```

```python
def make_dashboard(x, gdp_data, unemployment):
    
    plt.plot(year, unemployment, label='Unemployment')
    plt.plot(year, gdp_data, label='GDP Change')
    plt.title('US GDP Change and Unemployment Rate')
    plt.xlabel('Year')
    plt.ylabel('Unemployment & GDP Change')
    plt.legend(loc='upper right')
    plt.show()
    
```

```python
make_dashboard(year, gdp_data, unemployment)
```
